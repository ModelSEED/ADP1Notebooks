# AI Checking Gene Associations to Reactions

```python
%run util.py
pubmod = MSModelUtil.from_cobrapy("data/PublishedModel.XML")
rxndata = util.load("reactiondata")
genedata = util.load("genedata_enhanced")
rxn_mapping = {}
for rxn in pubmod.model.reactions:
    if rxn.id in rxndata:
        rxn_mapping.setdefault(rxn.id, {})
        genelist = {}
        for geneid in rxndata[rxn.id]["Match genes"]:
            if geneid in genedata:
                genelist[geneid] = genedata[geneid]
        for geneid in rxndata[rxn.id]["MS only genes"]:
            if geneid in genedata:
                genelist[geneid] = genedata[geneid]
        for geneid in rxndata[rxn.id]["Model only genes"]:
            if geneid in genedata:
                genelist[geneid] = genedata[geneid]
        for geneid in genelist:
            if "Reactions" in genelist[geneid]:
                del genelist[geneid]["Reactions"]
            if "Proteomics" in genelist[geneid]:
                del genelist[geneid]["Proteomics"]
            rxn_mapping[rxn.id][geneid] = util.evaluate_reaction_gene_association(rxn, genelist[geneid])
        util.save("rxn_gene_mapping",rxn_mapping)
```
